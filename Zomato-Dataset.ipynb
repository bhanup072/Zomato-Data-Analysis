## Information about all the columns in the dataset
1. **`url`**: The URL associated with the restaurant.
2. **`address`**: The address of the restaurant's physical location.
3. **`name`**: The name of the restaurant.
4. **`online_order`**: Indicates whether online ordering is available at the restaurant.
5. **`book_table`**: Indicates whether customers can book a table online at the restaurant.
6. **`rate`**: The rating of the restaurant (could be on a scale of 1 to 5 or any other rating scale).
7. **`votes`**: The number of votes or ratings the restaurant has received.
8. **`phone`**: The contact phone number of the restaurant.
9. **`location`**: The geographical location or area where the restaurant is situated.
10. **`rest_type`**: The type or category of the restaurant (e.g., cafe, fine dining, bar).
11. **`dish_liked`**: Dishes that are popular or well-liked at the restaurant.
12. **`cuisines`**: The types of cuisines offered by the restaurant.
13. **`approx_cost(for two people)`**: The approximate cost for two people dining at the restaurant.
14. **`reviews_list`**: A list of reviews or comments about the restaurant.
15. **`menu_item`**: The menu items offered by the restaurant.
16. **`listed_in(type)`**: The type of listings (e.g., Buffet, Delivery) the restaurant is listed under.
17. **`listed_in(city)`**: The city or region where the restaurant is listed.
